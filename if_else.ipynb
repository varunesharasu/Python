```python
# Percentage Calculator

per=float(input())

if(per>=85 and per<=100):
    print("You have scored {} percentage with Distinction".format(per))
elif(per>=60 and per<=84):
    print("First class")
elif(per>=50 and per<=59):
    print("Second class")
elif(per<=49 and per>=35):
    print("Pass")
else:
    print("Fail")
```

```python
""" 
write aprogram to check the given input forms which triangle
1. Equilateral
2. Isoceles
3. Scalene
4. Invalid Triangle
"""

a=int(input())
b=int(input())
c=int(input())

if(a+b<c or a+c<b or b+c<a):
    print("Invalid Triangle")
else:
    if(a==b and b==c and c==a):
        print("Equilateral Triangle")
    elif(a==b or a==c or b==c):
        print("Isoceles Triangle")
    else:
        print("Scalene Triangle")
```

```python
''' 
Calculate final bill amount
purchase amount 0 to 5000 no discount
purchase amount above 5000 to 10000 is 10% discount
purchase amount above 10000 to 20000 is 15% discount
purchase amount above 20000 to 30000 is 20% discount
purchase amount above 30000 flat 30% discount
'''

a=float(input())

if(a>=0 and a<=5000):
    print("No Discount")
elif(a>5000 and a<=10000):
    dis=a*0.1
    ba=a-dis
    print("10% Discount and your final amount is: ",ba)
elif(a>10000 and a<=20000):
    dis=a*0.15
    ba=a-dis
    print("15% Discount and your final amount is: ",ba)
elif(a>20000 and a<=30000):
    dis=a*0.2
    ba=a-dis
    print("20% Discount and your final amount is: ",ba)
elif(a>30000):
    dis=a*.30
    ba=a-dis
    print("30% Discount and your final amount is: ",ba)
else:
    print("InValid Amount")
```

```python
# Check the number is positive, negative or zero using nested condition statement

a=int(input())

if(a!=0):
    if(a>0):
        print("Positive number")
    else:
        print("Negative number")
else:
    print("entered number is zero")
```
